## Concatenate two data frames using the following conditions
**Create two data frames using the following two dictionaries.**

GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}

japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

```python
germanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'price': [29995, 23600, 61500 , 58900]}
cars_df1 = pd.DataFrame(germanCars)
cars_df2 = pd.DataFrame(japaneseCars)
```

```python
cars_df1['origin_city'] = ['jermany' if i == 0 else '' for i in range(len(cars_df1))]
cars_df2['origin_city'] = ['japan' if i == 0 else '' for i in range(len(cars_df2))]
```

```python
cars_df = pd.concat([cars_df1, cars_df2])
cars_df = cars_df[['origin_city', 'Company', 'price']]
cars_df.columns = ['', 'company', 'price']
cars_df
```
